# ESCI Dataset Exploration

This notebook explores the Amazon Shopping Queries (ESCI) Dataset, performing initial data loading, basic statistics analysis, and data merging.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

sns.set_theme(style="whitegrid")

# Set path to raw data
DATA_DIR = "../data/raw"
```

## 1. Load Datasets

```python
examples_path = os.path.join(DATA_DIR, "shopping_queries_dataset_examples.parquet")
products_path = os.path.join(DATA_DIR, "shopping_queries_dataset_products.parquet")
sources_path = os.path.join(DATA_DIR, "shopping_queries_dataset_sources.csv")

examples_df = pd.read_parquet(examples_path)
products_df = pd.read_parquet(products_path)
sources_df = pd.read_csv(sources_path)

print(f"Examples Dataset Shape: {examples_df.shape}")
print(f"Products Dataset Shape: {products_df.shape}")
print(f"Sources Dataset Shape: {sources_df.shape}")
```

## 2. Check Columns and Data

```python
print("Examples Columns:", examples_df.columns.tolist())
display(examples_df.head(10))
```

```python
print("Products Columns:", products_df.columns.tolist())
display(products_df.head(10))
```

## 3. Analyze ESCI Label Distribution

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=examples_df, x='esci_label', order=['E', 'S', 'C', 'I'])
plt.title('Distribution of ESCI Labels')
plt.xlabel('ESCI Label')
plt.ylabel('Count')
plt.show()

print(examples_df['esci_label'].value_counts(normalize=True) * 100)
```

## 4. Check Missing Values

```python
print("Missing values in Examples:\n", examples_df.isnull().sum())
print("\nMissing values in Products:\n", products_df.isnull().sum())
```

## 5. Merge Datasets

```python
merged_df = pd.merge(examples_df, products_df, on=['product_id', 'product_locale'], how='left')

print(f"Merged Dataset Shape: {merged_df.shape}")
print("Merged Dataset Columns:", merged_df.columns.tolist())
display(merged_df.head(10))
```

## 6. Save Merged Dataset

```python
output_dir = "../data/processed"
os.makedirs(output_dir, exist_ok=True)
output_path = os.path.join(output_dir, "merged_dataset.csv")

merged_df.to_csv(output_path, index=False)
print(f"Merged dataset saved successfully to {output_path}")
```
